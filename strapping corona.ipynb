```python
from bs4 import BeautifulSoup
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
```

```python
import requests
response = requests.get(url)
response
```

```python
soup = response.text
soup = BeautifulSoup(soup,"lxml") 
soup
```

```python
table = soup.class="table table-bordered table-hover main_table_countries dataTable no-footer
```

```python
table
```

```python
table.find_all('tr')
```

```python
tags = table.find_all('tr')
```

```python
tags[1].text
```

```python
data = []
for i in tags:
    data.append(i.text)
```

```python
data
```

```python
for i in data:
    print(i.split("\n")[1:-11])
```

```python
final_output=[]
for i in data:
    output = i.split('\n')[1:-1]
    final_output.append(output)
```

```python
final_output
```

```python
import csv
with open('corona.csv','a') as f:
    x = csv.writer(f)
    for i in final_output:
        x.writerow(i)
```
