```python
from bs4 import BeautifulSoup
url = BeautifulSoup('http://www.nepalstock.com/floorsheet', 'html.parser')
```

```python
import requests
response = requests.get(url)
response
```

```python
soup = response.text
soup = BeautifulSoup(soup,"lxml")
soup
```

```python
table = soup.table
```

```python
table.find_all('tr')
```

```python
tags = table.find_all('tr')
```

```python
tags[2].text
```

```python
data = []
for i in tags:
    data.append(i.text)
```

```python
data
```

```python
data = data [1:-3]
```

```python
data
```

```python
for i in data:
    print(i.split("\n")[0:-1])
```

```python
final_output=[]
for i in data:
    output = i.split('\n')[1:-1]
    final_output.append(output)
```
